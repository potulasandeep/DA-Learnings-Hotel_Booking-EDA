# Data Cleaning

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.agent.fillna('agent not available',inplace=True)
df.company.fillna('company not available',inplace=True)
df.dropna(subset = ['company','children'],inplace=True)
df.info()
```

```python
df.sample(10)
```

# Reservation Status Count

```python
import matplotlib.pyplot as plt
import seaborn as sn
%matplotlib inline
df.is_canceled.value_counts()
#df_res = df.groupby('is_canceled',as_index=False)['is_canceled'].count()
plt.bar(['not canceled','canceled'],df.is_canceled.value_counts())
plt.title('Reservation Status Count')
```

```python
df_res = df.groupby('is_canceled',as_index=False)['hotel'].count()
#df_res.is_canceled = ['not canceled','canceled']
df_res
ax = sn.barplot(x='is_canceled',y='hotel',data=df_res,label=['not canceled','canceled'])
ax.set_ylabel('count')
ax.set_xlabel('status')
ax.set_title('Reservation status count')
ax.legend()
```

# Top 10 Countries with reservation canceled

```python
df.sample(10)
```

```python
df_country = df.groupby(['country','is_canceled'],as_index=False)['hotel'].count().sort_values('hotel',ascending=False)
df_country = df_country[df_country.is_canceled == 1]
df_country = df_country.head(10)
df_country = df_country.drop(['is_canceled'],axis=1)
plt.pie(df_country.hotel,labels=df_country.country,radius=2,autopct="%0.2f%%",shadow=True)
plt.show()
```

```python
df_cou = df[df.is_canceled == 1]
df_top_10 = df_cou['country'].value_counts()[:10]
plt.pie(df_top_10,labels=df_top_10.index,autopct='%0.2f%%',radius = 2)
plt.show()
```

# Count of no.of stays on weekend_nights VS  stays on_week_nights

```python
#count of no .of stays on weekend nights
df_stay_weekend = df.groupby(['stays_in_weekend_nights'],as_index=False)['stays_in_weekend_nights'].value_counts()
ax =sn.barplot(x='stays_in_weekend_nights',y='count',data=df_stay_weekend)
plt.yscale('log')
ax.set_title('Count of no.of stays on weekend_nights')
ax.set_ylabel("log(count)")
```

```python
#count of no.of stays on week nights
df_stay_week = df.groupby('stays_in_week_nights',as_index=False)['stays_in_week_nights'].value_counts()
df_stay_week
ax=sn.barplot(x='stays_in_week_nights',y='count',data=df_stay_week)
plt.yscale('log')
plt.ylabel('Log(count)')
plt.title("count of no.of stays on week nights")
plt.xticks(rotation=60)
plt.show()
```

# Hotel Type Preferred by Adults

```python
df_adults = df.groupby(['hotel','adults'],as_index=False)['adults'].count()
df_adults_city = df_adults[df_adults.hotel == 'City Hotel']
df_adults_resort = df_adults[df_adults.hotel == 'Resort Hotel']
adults = [df_adults_city.adults.sum(),df_adults_resort.adults.sum()]
plt.bar(['City Hotel','Resort Hotel'],adults)
plt.ylabel('number of adults')
plt.title('Hotel Type Preferred by Adults')
#ax = sn.barplot(x='hotel',y='count',data=)
#plt.title("City Hotel preffered by no.of adults")
```
